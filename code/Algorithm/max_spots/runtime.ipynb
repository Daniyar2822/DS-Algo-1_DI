```python
# Import the gs_match function from the gs_places module and other necessary modules
from gs_places import gs_match
import time
import matplotlib.pyplot as plt
import random


# Define a function to generate large amounts of preferences for students and professors
def generate_large_preferences(num_students, num_professors):
    students_preferences = {}
    prof_preferences = {}

    for student_id in range(num_students):
        student_pref = random.sample(range(num_professors), num_professors)
        students_preferences[student_id] = student_pref

    for prof_id in range(num_professors):
        prof_pref = random.sample(range(num_students), num_students)
        prof_preferences[prof_id] = prof_pref

    return students_preferences, prof_preferences

# Define gs match implementation, accounting for maximum capacity of professors
def gs_match(students_preferences, prof_preferences, places):
    matches = {}
    remaining_places = places.copy()
    unmatched_students = []

    free_student = list(students_preferences.keys())

    while free_student:
        student = free_student.pop(0)
        if students_preferences[student]:
            prof = students_preferences[student].pop(0)
            if prof not in matches:
                matches[prof] = [student]
                remaining_places[prof] -= 1
            elif remaining_places[prof] > 0:
                matches[prof].append(student)
                remaining_places[prof] -= 1
            else:
                prof_pref_list = prof_preferences[prof]
                current_match = matches[prof][0]

                if prof_pref_list.index(current_match) > prof_pref_list.index(student):
                    matches[prof] = [student]
                    free_student.append(current_match)
                    remaining_places[prof] -= 1
                else:
                    free_student.append(student)
        else:
            unmatched_students.append(student)

    return matches, unmatched_students

# Lists to store input sizes and runtimes
input_sizes = []
runtimes = []

# Define the initial input size, the maximum amount of students, and the iteration step size
start_num_students = 1
start_num_professors = 1
max_num_students = 5000
step = 100



# Generate preferences and adjust places 
num_students = start_num_students
num_professors = start_num_professors
while num_students <= max_num_students:
    # Generate large preferences
    students_preferences, prof_preferences = generate_large_preferences(num_students, num_professors)

    # Generate random places for professors
    places = {professor: random.randint(1, num_students) for professor in range(num_professors)}

    # Test the runtime of gs_match
    start_time = time.time()
    # Run the actual matching process on the above defined preferences
    gs_match(students_preferences, prof_preferences, places)
    end_time = time.time()
    runtime = end_time - start_time

    # Append input size and runtimes to the lists
    input_sizes.append(num_students)
    runtimes.append(runtime)

    # Increase the input size on both side by the defined step size
    num_students += step
    num_professors += step

# Plot the results
'''
# Plot the results
plt.plot(input_sizes, runtimes, marker='o')
plt.title('Runtime of Gale-Shapley Algorithm vs Input Size')
plt.xlabel('Input Size (Number of Students/Professors)')
plt.ylabel('Runtime (seconds)')
plt.grid(True)
plt.show()
'''
```
